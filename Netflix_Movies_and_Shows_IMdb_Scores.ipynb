```python
df['description'].fillna('Unavailable', inplace=True)
```

```python
# Cleaned null values from description column and checking for null values in other columns
df.isnull().sum()
```

```python
# Decided to fill the null values on the age_certification as PG-17 which describes the category as Parental Guidance. Some material may not be suitable for children under 17.
df['age_certification'].fillna('PG-17', inplace=True)
```

```python
# Cleaned null values from age_certification column and checking for null values in other columns
df.isnull().sum()
```

```python
# Decided to fill the null values on the imdb_votes as 0
df['imdb_votes'].fillna(0, inplace=True)
```

```python
# All null values cleared. Now encoding the object data types to number forms for machine learning
from sklearn.preprocessing import OrdinalEncoder
encode = OrdinalEncoder()
```

```python
# Loop through columns with object dtype and print their names along with their number of unique values
for i in df.select_dtypes(include=["object"]).columns:
  print(i,df[i].nunique())
```

```python
# Encoded title and age_certification with OrdinalEncoder
df["title"] = encode.fit_transform(df[["title"]])
df["age_certification"] = encode.fit_transform(df[["age_certification"]])
df["description"] = encode.fit_transform(df[["description"]])
```

```python
# Displaying the encoded results
df
```

```python
# Checking the possibilities of convertion in type column
df['type'].unique()
```

```python
# Replacing MOVIE as 1 and SHOW as 0
df["type"]=df["type"].replace({"MOVIE":1,"SHOW":0})
df
```

```python
# Loop through columns with object dtype and print their names along with their number of unique values
for i in df.select_dtypes(include=["object"]).columns:
  print(i,df[i].nunique())
```

```python
# Converting id and imdb_id into numbers
to_remove_id = ['tm', 'ts']
for i in to_remove_id:
    df['id'] = df['id'].str.replace(i, '', regex=False)
df['imdb_id'] = df['imdb_id'].str.replace('tt', '', regex=False)
df
```

```python
# Checking if any duplicates in id column.
df[df['id'].duplicated(keep='first')]
```

```python
# Adding 1 at the end of duplicated id's to make it unique.
df1 = df['id'].duplicated(keep='first')
df.loc[df1, 'id'] = df.loc[df1, 'id'] + '1'
# Loop through columns with object dtype and print their names along with their number of unique values
for i in df.select_dtypes(include=["object"]).columns:
  print(i,df[i].nunique())
```

```python
# Converted All object data type into integer and all data in the data frame is in numeric format
df['id'] = df['id'].astype('Int64')
df['imdb_id'] = df['imdb_id'].astype('Int64')
```

```python
df.info()
```

```python
# Performing T test for all columns to check data is properly distributed
from scipy import stats
sample = df['id'].sample(frac=0.03)
```

```python
t_test,p_value = stats.ttest_1samp(sample,popmean=df['id'].mean())
```

```python
if p_value<0.05:
  print("Reject the null hypothesis, data is not properly distributed")
else:
  print("Accept the null hypothesis, data is properly distributed")
```

```python
# Mathematically we can perform 30 sample t test to check is properly distributed
for i in range(31):
  sample = df['id'].sample(frac=0.03)
  t_test,p_value = stats.ttest_1samp(sample,popmean=df['id'].mean())
  if p_value<0.05:
    print("Reject the null hypothesis, data is not properly distributed")
  else:
    print("Accept the null hypothesis, data is properly distributed")
```
